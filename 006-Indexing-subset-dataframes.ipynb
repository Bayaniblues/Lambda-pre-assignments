As you saw in the lecture, sometimes we want to just use part of a DataFrame in our analysis, especially if we're working with larger datasets. We can get a subset of the data by selecting a range of rows and columns.

### 1.8 - Filtering DataFrames

Filtering is when we subset a DataFrame using various criteria or conditions. We'll continue working with the same dataset and try out some basic filtering.

In the following cell, return all entries where the `size` column has a value greater than 2. In another cell, create a new DataFrame consisting only entries that contain Friday and call it `df_friday`.

```
# Return all the entries where the size is greater than 2.
greater_than = lambda Column_name, size : df[Column_name] > size
greater_than('size', 2)
```

```
# Create a DataFrame consisting only of data for Fridays. 
# Hint: You should have 19 rows.
equal_to = lambda Column_name, day : df[Column_name] == day 
mask = equal_to('day', 'Fri')
df_friday = df[equal_to('day', 'Fri')]
df_friday
```
